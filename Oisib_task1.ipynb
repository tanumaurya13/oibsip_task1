Iris Flower Classification

```python
# Import Packages

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
# Loading data
df = pd.read_csv('iris.data', names=columns)
df.head(150)
```

```python
df.describe()
```

```python
sns.scatterplot(x='Sepal length',y='Sepal width',hue='Class_labels', data=df)
# Add labels and title
plt.xlabel('sepal length')
plt.ylabel('Sepal width')
plt.title('Iris dataset - Sepal length vs. Sepal width')
# Show the plot
plt.show()
```

```python
sns.scatterplot(x='Petal length',y='Petal width',hue='Class_labels', data=df)
# Add labels and title
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.title('Iris dataset - Petal length vs. Petal width')
# Show the plot
plt.show()
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(10, 10))
sns.violinplot(x="Class_labels", y="Sepal length", data=df, ax=ax1)
sns.violinplot(x="Class_labels", y="Sepal width", data=df, ax=ax2)
sns.violinplot(x="Class_labels", y="Petal length", data=df, ax=ax3)
sns.violinplot(x="Class_labels", y="Petal width", data=df, ax=ax4)
```

```python
# Visualize the whole dataset
sns.pairplot(df, hue='Class_labels')
```

```python
# Separate features and target
data = df.values
X = data[:,0:4]
print(X)
```

```python
Y = data[:,4]
print(Y)
```

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(X_test)
```

1) SVC



```python
# Support vector machine algorithm
from sklearn.svm import SVC

mdsvc = SVC()
mdsvc.fit(X_train, y_train)
```

```python
# Predict from the test dataset
prediction1 = mdsvc.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction1)*100
```

2) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
mdlr=LogisticRegression()
mdlr.fit(X_train,y_train)
```

```python
prediction2=mdlr.predict(X_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction2)*100
```

3)Random forest

```python
from sklearn.ensemble import RandomForestClassifier
mdrf=RandomForestClassifier()
mdrf.fit(X_train,y_train)
```

```python
prediction3=mdrf.predict(X_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction3)*100
```

4) KNN

```python
from sklearn.neighbors import KNeighborsClassifier
mdknn=KNeighborsClassifier()
mdknn.fit(X_train,y_train)
```

```python
prediction4=mdknn.predict(X_test)
# Calculate the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction4)*100
```
